```python
def augment_data(group):
    augmented_data = []
    for i in range(1, 11):
        for j in range(len(group)-i):
            ayah_ens = ""
            for k in range(j, j+i+1):
                ayah_ens += group.iloc[k]["ayah_en"]
                if k != j+i:
                    ayah_ens += " "
            augmented_data.append({
                "surah_no": group.iloc[j]["surah_no"],
                "ayah_no_surah": f"{group.iloc[j]['ayah_no_surah']}-{group.iloc[j+i]['ayah_no_surah']}",
                "ayah_en": ayah_ens
            })
    return pd.DataFrame(augmented_data)
```

```python
df_augmented = dfSelected.groupby("surah_no", group_keys=False).apply(augment_data)
df_augmented.head()
```

```python
import pandas as pd
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer

modelEmbedding = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
from tqdm import tqdm

df = pd.read_csv("The Quran Dataset Augmented.csv")

embeddings = []

for ayah in tqdm(df["ayah_en"].tolist(), desc="Encoding embeddings", unit="ayah"):
    embeddings.append(modelEmbedding.encode(ayah, convert_to_numpy=True))

embeddings = np.array(embeddings)

df["ayah_embedding"] = embeddings.tolist()
```

```python
dimension = embeddings.shape[1] 
index = faiss.IndexFlatIP(dimension)

index.add(embeddings)

faiss.write_index(index, "faiss_index.bin")
df.to_csv("data_with_embeddings.csv", index=False) 
```

```python
import pandas as pd

df = pd.read_csv("The Quran Dataset Augmented.csv")
```

```python
df[['first_ayah_no_surah', 'last_ayah_no_surah']] = df['ayah_no_surah'].str.split('-', expand=True)
df['last_ayah_no_surah'] = df['last_ayah_no_surah'].fillna(df['first_ayah_no_surah'])
df = df.drop(columns=['ayah_no_surah'])
df.to_csv("The Quran Dataset Augmented.csv", index=False)
```

```python
import google.generativeai as genai
```

```python
GEMINI_API_KEY = "apikey"
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-2.0-flash")
```

```python
query = "What is the importance of patience?"

prompt = f""" 
    Role: You are a helpful assistant tasked with identifying keywords in the user's query.
    Goal: Given user's query, provide the keywords to help find answer in corpus.
    User Question: {query}
    Keywords:
"""
```

```python
response = model.generate_content(prompt)
response.text
```

```python
encoded_text = modelEmbedding.encode(response.text, convert_to_numpy=True).reshape(1, -1)
```

```python
import os

FAISS_INDEX_PATH = "faiss_index.bin"
VECTOR_DIM = 384

if os.path.exists(FAISS_INDEX_PATH):
    index = faiss.read_index(FAISS_INDEX_PATH)
else:
    raise FileNotFoundError(f"FAISS index file '{FAISS_INDEX_PATH}' not found!")
```

```python
distances, indices = index.search(encoded_text, 10)

filtered_indices = [indices[0][i] for i in range(len(indices[0])) if distances[0][i] > 0.1]

retrieved_data = [df.iloc[i] for i in filtered_indices if i < len(df)]

retrieved_data_df = pd.DataFrame(retrieved_data)

def remove_subsets(group):
    filtered_rows = []
    selected_rows = []

    group = group.sort_values(['first_ayah_no_surah', 'last_ayah_no_surah'], ascending=[True, False]).reset_index(drop=True)

    for i, row in group.iterrows():
        first, last = row['first_ayah_no_surah'], row['last_ayah_no_surah']
        
        if any(f <= first and l >= last for f, l in filtered_rows):
            continue
        
        filtered_rows.append((first, last))
        selected_rows.append(row)

    return pd.DataFrame(selected_rows)

df_filtered = retrieved_data_df.groupby('surah_no', group_keys=False).apply(remove_subsets).reset_index(drop=True)

df_filtered
```
